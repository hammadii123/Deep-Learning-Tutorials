```python
# Scalar Step Function
import math
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt


def step_function(x):
    return 1 if x >= 0 else 0

# Vectorized Step Function (TensorFlow)
def step_tf(x):
    return tf.cast(x >= 0, tf.int32)

# Test Input
inputs1 = [-3.0, -1.0, 0.0, 2.0, 5.0]
outputs1 = [step_function(x) for x in inputs1]
print("Step Function Output:", outputs1)

# TensorFlow Version
inputs1_tf = tf.constant(inputs1)
outputs1_tf = step_tf(inputs1_tf)
print("Step Function TF:", outputs1_tf.numpy())

# Visualization
x_vals = np.linspace(-5, 5, 200)
y_vals = [step_function(x) for x in x_vals]

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label="Step Function")
plt.title("Step Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def relu(x):
    return x if x > 0 else 0

inputs = [-4.0, -2.0, 0.0, 3.0, 5.0]
outputs = [relu(i) for i in inputs]
print("ReLU Output:", outputs)
x = np.linspace(-5, 5, 100)
y = [relu(i) for i in x]

plt.plot(x, y)
plt.title("ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Task3
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

inputs = [-2.0, 0.0, 1.0, 3.0]
outputs = [round(sigmoid(i), 3) for i in inputs]
print("Sigmoid Output:", outputs)

x = np.linspace(-5, 5, 100)
y = [sigmoid(i) for i in x]

plt.plot(x, y)
plt.title("Sigmoid")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Task4
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

inputs = [-1.5, 0.0, 1.0, 2.0]
outputs = [round(tanh(i), 3) for i in inputs]
print("Tanh Output:", outputs)
x = np.linspace(-5, 5, 100)
y = [tanh(i) for i in x]

plt.plot(x, y)
plt.title("Tanh Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
def leaky_relu(x, alpha=0.1):
    return x if x >= 0 else alpha * x

inputs = [-3.0, -1.0, 0.0, 2.0]
outputs = [leaky_relu(i) for i in inputs]
print("Leaky ReLU Output (alpha=0.1):", outputs)
x = np.linspace(-5, 5, 100)
y1 = [leaky_relu(i, alpha=0.1) for i in x]
y2 = [leaky_relu(i, alpha=0.3) for i in x]

plt.plot(x, y1, label="alpha=0.1")
plt.plot(x, y2, label="alpha=0.3")
plt.title("Leaky ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task6
def softmax(x):
    exp_vals = [math.exp(i) for i in x]
    total = sum(exp_vals)
    return [round(j / total, 3) for j in exp_vals]

inputs = [1.0, 2.0, 3.0]
outputs = softmax(inputs)
print("Softmax Output:", outputs)

labels = ['1.0', '2.0', '3.0']
plt.bar(labels, outputs)
plt.title("Softmax Output")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
