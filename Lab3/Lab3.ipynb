```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 2: Load the dataset
data = pd.read_csv("Salary_dataset.csv")  
X = data[['YearsExperience']]
y = data['Salary'] 

# Step 3: Train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Make predictions
predicted_salary_3 = model.predict([[3]])
predicted_salary_10 = model.predict([[10]])

# Step 5: Print predictions
print("Predicted salary for 3 years of experience:", predicted_salary_3[0])
print("Predicted salary for 10 years of experience:", predicted_salary_10[0])

#View model stats
print("Slope (Coefficient):", model.coef_[0])
print("Intercept (Bias):", model.intercept_)

# (Optional) Mean Squared Error
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Load the dataset
df = pd.read_csv('Salary_dataset.csv')

# Step 2: Add the outlier data point
outlier = pd.DataFrame({'YearsExperience': [1.5], 'Salary': [150000]})
df = pd.concat([df, outlier], ignore_index=True)

# Step 3: Split the data into X (input) and y (output)
X = df[['YearsExperience']]
y = df['Salary']

# Step 4: Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 5: Plot the regression line with the outlier
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.title('Salary vs Experience (With Outlier)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Predict salary for 6 years of experience
predicted_salary = model.predict([[6]])
print(f"Predicted salary for 6 years of experience: ₹{predicted_salary[0]:.2f}")

# Step 7: Print the slope and intercept
print(f"Slope (Coefficient): ₹{model.coef_[0]:.2f}")
print(f"Intercept (Bias): ₹{model.intercept_:.2f}")

# Step 8: Calculate and print Mean Squared Error
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# Slope (Coefficient)
print("Slope (Coefficient):", model.coef_[0])

# Intercept (Bias)
print("Intercept (Bias):", model.intercept_)
```
